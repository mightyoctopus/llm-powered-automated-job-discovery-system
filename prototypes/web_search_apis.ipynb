# SerpAPI Web Search

```python
import os
import serpapi
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
SERP_API_KEY = os.getenv("SERP_API_KEY")
```

```python
client = serpapi.Client(api_key=SERP_API_KEY)

results = client.search({
    "engine": "google", 
    "google_domain": "google.com",
    "hl": "en", 
    "q": 'intitle:(hiring OR careers OR jobs) (LLM engineering OR AI agents OR agentic systems OR RAG pipeline OR LoRA QLoRA OR fine-tuning OR AI integration) (remote OR worldwide OR Asia OR APAC) (inurl:greenhouse.io OR inurl:lever.co OR inurl:workable OR inurl:careers)'
})
```

```python
results["organic_results"][1]
```

# Exa Web Search

```python
from exa_py import Exa 
```

```python
EXA_API_KEY = os.getenv("EXA_API_KEY")

if not EXA_API_KEY:
    raise ValueError("Exa API key not found in env variables.")
```

```python
exa_client = Exa(api_key=EXA_API_KEY)
```

```python
results = exa_client.search(
    query="Company careers pages hiring interns or early-career LLM engineers for RAG pipelines, LoRA/QLoRA fine-tuning, agentic AI, or AI-driven app integration; remote-friendly roles accepting applicants from Asia/APAC or global",
    type="auto",
    contents={"highlights": {"max_characters": 4000}}
)
```

```python
len(results.results)
```
